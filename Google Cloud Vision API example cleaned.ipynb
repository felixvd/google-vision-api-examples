```python
# This is from the HowTo for Detecting Text: https://cloud.google.com/vision/docs/detecting-text
def detect_text(path):
    """Detects text in the file."""
    vision_client = vision.Client(project=GOOGLE_CLOUD_PROJECT)

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision_client.image(content=content)

    texts = image.detect_text()
    print('Texts:')
    return texts, image
```

```python
texts, image = detect_text('receipt1.jpg')
```

```python
for text in texts:
    print(text.description)
```
